```python
def find_text_split(text):
    if re.search('[0-9]-[0-9]',text):
        span = re.search('[0-9]-[0-9]',text).span()
        text = re.sub('[0-9]-[0-9]',text[span[0]]+'\n-'+text[span[0]+2],text)
    return text
```

```python
def wrap_text(text):
    temp = []
    skip = False
    for i in range(len(text)):
        if skip:
            skip = False
            pass
        else:
            if text[i][-1] == ' ':
                temp.append(text[i]+text[i+1])
                if text[i+1][-1] != ' ':
                    skip = True
            else:
                temp.append(text[i])
    return temp
```

```python
text = []
regex_sections = {'[(]1[)][(]a[)]':'(1)(a)',
                '[(]2[)][(]a[)]':'(2)(a)',
                '[(]2[)][(]b[)]':'(2)(b)',
                '[(]2[)][(]c[)]':'(2)(c)',
                '[(]2[)][(]d[)]':'(2)(d)'}
sections = [v for v in regex_sections.values()]
for i in range(pdfReader.numPages):
    page = pdfReader.getPage(i).extractText()
    page = find_text_split(page)
    page = re.sub('Page [0-9]+ of [0-9]+','',page)
    for k,v in regex_sections.items():
        page = re.sub(k,'\n'+v,page)
    page = re.sub('AXLE WEIGHT','AXLE WEIGHT\n',page)
    page = re.sub('-*Local:','\nLocal:',page)
    page = re.sub('All Other Roads in Improvement Districts\n','All Other Roads in Improvement Districts',page)
    page = re.split('\n',page)
#     page = re.split('-',page)
    page = [x for x in page if (x != '')]
    page = [x for x in page if (x != '-')]
    text.extend(page)
#     text.append(page)
text
```

### Structure

```python
def create_structure(text,sections):
    structure = {}
    for i in range(len(sections)):
        for j in range(len(text)):
            if i < len(sections)-1:
                if sections[i] in text[j]:
                    section_start = j
                    section_end = j
                    for k in range(j+1,len(text)):
                        if sections[i+1] in text[k]:
                            section_end = k
                            break
                    structure.setdefault(sections[i],text[section_start:section_end])
            else:
                if sections[i] in text[j]:
                    section_start = j
                    structure.setdefault(sections[i],text[section_start:])
    return structure
```

```python
def parse2a(text):
    new_text = []
    for x in text:
        if re.search('^-*[0-9]+[A-Z][A-Z]',x):
            span = re.search('^-*[0-9]+[A-Z][A-Z]',x).span()
            new_text.append(re.sub('^-*[0-9]+[A-Z][A-Z]',x[span[0]:span[1]-1]+'\t'+x[span[1]-1],x))
        elif re.search('^-[0-9]+[A-Z]',x):
            span = re.search('^-[0-9]+[A-Z]',x).span()
            new_text.append(re.sub('^-[0-9]+[A-Z]',x[span[0]:span[1]-1]+'\t'+x[span[1]-1],x))
        elif re.search('^-[0-9]+[.\s]',x):
            span = re.search('^-[0-9]+[.\s]',x).span()
            if span[1]>6:
                new_text.append(re.sub('^-[0-9]+[.\s]',x[span[0]:4]+'\t'+x[(4):span[1]],x))
            else:
                new_text.append(re.sub('^-[0-9]+[.\s]',x[span[0]:span[1]-2]+'\t'+x[(span[1]-2):span[1]],x))
        else:
            new_text.append(x)
    new_text = wrap_text(new_text)
    new_text = [x for x in new_text if x != '-']
    new_text = [x[1:] if x[0] == '-' else x for x in new_text]
    new_text = [x for x in new_text if re.search('[0-9]$',x)]
    new_text = ['\t'+x if x[0] == ' ' else x for x in new_text]
    new_text = [re.sub('100$','\t100',x) for x in new_text]
    new_text = [re.sub('[1-9][0-9]$','\t'+x[-2:],x) for x in new_text]
    new_text = [x.split('\t') for x in new_text]
    return new_text
```

```python
def parse2c(text):
    new_text = []
    for i in range(len(text)):
        temp = text[i]
        for j in range(len(text)-i):
            if text[i][:6] == 'Local:':
                if text[i+j][:6] != 'Local:':
                    temp += text[i+j]
        if text[i][:6] == 'Local:':
            new_text.append(temp)
    new_text = [re.sub('Local:','Local\t',x) for x in new_text]
    new_text = [re.sub('100$','\t100',x) for x in new_text]
    new_text = [re.sub('[1-9][0-9]$','\t'+x[-2:],x) for x in new_text]
    new_text = [x.split('\t') for x in new_text]
    return new_text
```

```python
def parse_pdf(file):
    pdfFileObj = open(file,'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
    text = []
    regex_sections = {'[(]1[)][(]a[)]':'(1)(a)',
                    '[(]2[)][(]a[)]':'(2)(a)',
                    '[(]2[)][(]b[)]':'(2)(b)',
                    '[(]2[)][(]c[)]':'(2)(c)',
                    '[(]2[)][(]d[)]':'(2)(d)'}
    sections = [v for v in regex_sections.values()]
    for i in range(pdfReader.numPages):
        page = pdfReader.getPage(i).extractText()
        page = find_text_split(page)
        page = re.sub('Page [0-9]+ of [0-9]+','',page)
        for k,v in regex_sections.items():
            page = re.sub(k,'\n'+v,page)
        page = re.sub('AXLE WEIGHT','AXLE WEIGHT\n',page)
        page = re.sub('-*Local:','\nLocal:',page)
        page = re.sub('All Other Roads in Improvement Districts\n','All Other Roads in Improvement Districts',page)
        page = re.split('\n',page)
        page = [x for x in page if (x != '')]
        page = [x for x in page if (x != '-')]
        text.extend(page)
    structure = create_structure(text,sections[1:])
    headings = {'(2)(a)':['Highway','Description','Percentage Axle Weight'],
                '(2)(b)':['From Highway #','To Location','Percentage Axle Weight'],
                '(2)(c)':['Secondary Highway','Description','Percentage Axle Weight'],
                '(2)(d)':['Improvement District','Description','Percentage Axle Weight']}
    
    df_2a = pd.DataFrame(parse2a(structure['(2)(a)']),columns=headings['(2)(a)']),
    df_2b = pd.DataFrame(parse2b(structure['(2)(b)']),columns=headings['(2)(b)']),
    df_2c = pd.DataFrame(parse2c(structure['(2)(c)']),columns=headings['(2)(c)']),
    df_2d = pd.DataFrame(parse2d(structure['(2)(d)']),columns=headings['(2)(d)']),
    
    for df in [df_2a,df_2b,df_2c,df_2d]:
        df['Effective Date'] = [x for x in text if re.search('^EFFECTIVE',x)][0]
        
    outputs = {
        '(2)(a)':df_2a,
        '(2)(b)':df_2b,
        '(2)(c)':df_2c,
        '(2)(d)':df_2d,
    }
    details = {
        'effective date':[x for x in text if re.search('^EFFECTIVE',x)][0],
        'title':[x for x in text if re.search('^ROAD BAN ORDER',x)][0]
    }
    for key,output in outputs.items():
        output.to_csv(f"./outputs/{file[19:-4]}-{key}.csv")
        
    pdfFileObj.close()
    return outputs
```

```python
def parse_pdf_v2(file):
    pdfFileObj = open(file,'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
    text = []
    regex_sections = {'[(]1[)][(]a[)]':'(1)(a)',
                    '[(]2[)][(]a[)]':'(2)(a)',
                    '[(]2[)][(]b[)]':'(2)(b)',
                    '[(]2[)][(]c[)]':'(2)(c)',
                    '[(]2[)][(]d[)]':'(2)(d)'}
    sections = [v for v in regex_sections.values()]
    for i in range(pdfReader.numPages):
        page = pdfReader.getPage(i).extractText()
        page = find_text_split(page)
        page = re.sub('Page [0-9]+ of [0-9]+','',page)
        for k,v in regex_sections.items():
            page = re.sub(k,'\n'+v,page)
        page = re.sub('AXLE WEIGHT','AXLE WEIGHT\n',page)
        page = re.sub('-*Local:','\nLocal:',page)
        page = re.sub('All Other Roads in Improvement Districts\n','All Other Roads in Improvement Districts',page)
        page = re.split('\n',page)
        page = [x for x in page if (x != '')]
        page = [x for x in page if (x != '-')]
        text.extend(page)
    structure = create_structure(text,sections[1:])
    headings = {'(2)(a)':['Highway','Description','Percentage Axle Weight'],
                '(2)(b)':['From Highway #','To Location','Percentage Axle Weight'],
                '(2)(c)':['Secondary Highway','Description','Percentage Axle Weight'],
                '(2)(d)':['Improvement District','Description','Percentage Axle Weight']}

    df_2a = pd.DataFrame(parse2a_v2(structure['(2)(a)']),columns=headings['(2)(a)']),
    df_2b = pd.DataFrame(parse2b_v2(structure['(2)(b)']),columns=headings['(2)(b)']),
    df_2c = pd.DataFrame(parse2c_v2(structure['(2)(c)']),columns=headings['(2)(c)']),
    df_2d = pd.DataFrame(parse2d(structure['(2)(d)']),columns=headings['(2)(d)']),
    
    for df in [df_2a,df_2b,df_2c,df_2d]:
        df['Effective Date'] = [x for x in text if re.search('^EFFECTIVE',x)][0]
        
    outputs = {
        '(2)(a)':df_2a,
        '(2)(b)':df_2b,
        '(2)(c)':df_2c,
        '(2)(d)':df_2d,
    }
    
    details = {
        'effective date':[x for x in text if re.search('^EFFECTIVE',x)][0],
        'title':[x for x in text if re.search('^ROAD BAN ORDER',x)][0]
    }
    
    for key,output in outputs.items():
        output.to_csv(f"./outputs/{file[19:-4]}-{key}.csv")
        
    pdfFileObj.close()
    return outputs
```
